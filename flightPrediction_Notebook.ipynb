```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Test_Set.xlsx')
test_df.head()
```

```python
final_df = pd.concat([train_df,test_df])
final_df.head()
```

```python
final_df.info()
```

```python
## feature engineering
##Creating date , month and year separately
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
## removing the date from the arrival time as:
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute(s)']=final_df['Arrival_Time'].str.split(':').str[1]
final_df.head()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute(s)']=final_df['Arrival_Minute(s)'].astype(int)
final_df.info()
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head(1)
```

```python
final_df.isnull().sum()
```

```python
final_df['Dep_Time']=final_df['Dep_Time'].str.split(' ').str[0]
final_df['Dep_Time']
```

```python
final_df['Dep_Hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minute(s)']=final_df['Dep_Time'].str.split(':').str[1]
final_df.head()
```

```python
final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int)
final_df['Dep_Minute(s)']= final_df['Dep_Minute(s)'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
final_df.head(1)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4, 'nan':2})
```

```python
final_df.head()
```

```python
final_df
```

```python
final_df.drop('Route', axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.info()
```

```python
##converting the hours of Duration to minutes

final_df['Duration']=pd.to_timedelta(final_df['Duration']).dt.total_seconds() / 60
final_df['Duration']
```

```python
final_df['Duration'].astype(int)
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
## we will now perform label encoding on Airline ,Source,Destination  and for Additional_Info
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
